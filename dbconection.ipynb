```python
import psycopg2
from faker import Faker
import random
from datetime import datetime, time

dbname = "NotesStudyTest"
user = "postgres"
password = "1234"
host = "localhost"
port = "5432"

faker = Faker()

# Generar 100 datos de ejemplo para cada tabla
alumnos = [(i, faker.pyfloat(min_value=0, max_value=5), faker.random_int(min=1, max=8),
                faker.random_int(min=2000, max=2022), faker.random_element(["Masculino", "Femenino"]),
                faker.random_int(min=16, max=30)) for i in range(1, 101)]

apoyos = [(i, faker.city(), faker.random_int(min=1, max=6), faker.random_element(["Bus", "Trasmilenio", "Bicicleta","Carro","Moto","Taxi","Flota","Caminando","Ruta","Servicios de trasporte privados"]),
               faker.random_int(min=1, max=6), faker.random_element(["Sí", "No"]),
               faker.random_int(min=0, max=5), faker.random_int(min=5, max=120)) for i in range(1, 101)]

asistencias = [(i, faker.random_int(min=0, max=10), faker.random_int(min=0, max=5)) for i in range(1, 101)]

profesores = [(i, faker.name(), faker.random_int(min=1, max=10), faker.random_element(["A", "B", "C"]),
                   faker.random_int(min=10, max=40), faker.random_element(["Materia A", "Materia B", "Materia C","Materia D","Materia E","Materia F","Materia G","Materia H","Materia I","Materia J"]),
                   faker.job(), i) for i in range(1, 101)]

horarios = [time(random.randint(1, 10), random.randint(0, 59), 0) for _ in range(100)]

materias = [(i, faker.catch_phrase(), faker.random_int(min=1, max=4),
                     datetime.combine(datetime.today(), horarios[i - 1]).strftime('%I:%M%p'),
                     datetime.combine(datetime.today(), horarios[i - 1]).strftime('%I:%M%p'),
                     faker.random_element(["Lunes-Miércoles", "Martes-Jueves", "Viernes"]), i)
                    for i in range(1, 101)]

notas = [(i, i, random.uniform(0, 5), faker.random_element(["A", "B", "C"])
          , faker.random_int(min=0, max=10), faker.random_int(min=1, max=10)) for i in range(1, 101)]

try:
    connection = psycopg2.connect(
        dbname=dbname, user=user, password=password, host=host, port=port
    )
    print("Conexión exitosa a la base de datos")

    cursor = connection.cursor()


    # Insertar datos en las tablas
    cursor.executemany("INSERT INTO alumno VALUES (%s, %s, %s, %s, %s, %s)", alumnos)
    print("Alumnos insertados correctamente")
    cursor.executemany("INSERT INTO apoyoeconomico VALUES (%s, %s, %s, %s, %s, %s, %s, %s)", apoyos)
    print("Apoyos insertados correctamente")
    cursor.executemany("INSERT INTO asistencia VALUES (%s, %s, %s)", asistencias)
    print("Asistencias insertadas correctamente")
    cursor.executemany("INSERT INTO profesor VALUES (%s, %s, %s, %s, %s, %s, %s, %s)", profesores)
    print("Profesores insertados correctamente")
    cursor.executemany("INSERT INTO materia VALUES (%s, %s, %s, %s, %s, %s, %s)", materias)
    print("Materias insertadas correctamente")
    cursor.executemany("INSERT INTO nota VALUES (%s, %s, %s, %s, %s, %s)", notas)
    print("Notas insertadas correctamente")
    
    connection.commit()
    print("Datos de prueba insertados correctamente")

    cursor.close()

except psycopg2.Error as e:
    print("Error al conectarse a la base de datos:", e)
```

### Eliminar datos

```python
import psycopg2

# Parámetros de conexión
dbname = "NotesStudyTest"
user = "postgres"
password = "1234"
host = "localhost"
port = "5432"

try:
    connection = psycopg2.connect(
        dbname=dbname, user=user, password=password, host=host, port=port
    )
    print("Conexión exitosa a la base de datos")

    cursor = connection.cursor()

    # Eliminar los datos anteriores
    cursor.execute("DELETE FROM nota")
    cursor.execute("DELETE FROM asistencia")
    cursor.execute("DELETE FROM apoyoeconomico")
    cursor.execute("DELETE FROM materia")
    cursor.execute("DELETE FROM profesor")
    cursor.execute("DELETE FROM alumno")

    connection.commit()
    print("Datos de prueba reemplazados correctamente")

    cursor.close()

except psycopg2.Error as e:
    print("Error al conectarse a la base de datos:", e)
```
